#### Inicializa o retriever RAG

```python
retriever = get_rag_retriever()
llm = ChatGroq(model_name='llama-3.3-70b-versatile', temperature=0)
```

```python
def demo():
    print("\n=== SISTEMA DE TURISMO ===\n")

    # 1 Pergunta sobre Roteiro de Viagem
    print(">>> Pergunta 1: Roteiro de viagem")
    pergunta1 = "Quero um roteiro cultural de 3 dias em Paris"
    res1 = chains.run_itinerary(pergunta1, perfil="cultural")
    print("\n[Resposta Itinerary Chain]:\n", res1)

    # 2 Pergunta sobre Logística
    print("\n>>> Pergunta 2: Logística")
    pergunta2 = "Como chegar do aeroporto de Paris ao centro da cidade?"
    res2 = chains.run_logistics(pergunta2)
    print("\n[Resposta Logistics Chain]:\n", res2)

    # 3 Pergunta sobre informações locais com RAG
    print("\n>>> Pergunta 3: Informação local com contexto")
    pergunta3 = "Quais são os melhores restaurantes veganos em Paris?"
    
    # Busca no retriever do RAG para fornecer contexto
    contexto_docs = retriever.get_relevant_documents(pergunta3)
    contexto_texto = " ".join([doc.page_content for doc in contexto_docs])

    res3 = chains.run_localinfo(pergunta3, contexto=contexto_texto)
    print("\n[Resposta Local Info Chain]:\n", res3)

    # 4 Pergunta combinada (demonstrando contexto e fluxo)
    print("\n>>> Pergunta 4: Combinando Roteiro e info local")
    pergunta4 = "No meu roteiro cultural, inclua restaurantes veganos perto do Louvre"
    res4 = chains.run_itinerary(pergunta4, perfil="cultural")
    print("\n[Resposta Itinerary Chain + RAG]:\n", res4)

    print("\n=== FIM DA DEMONSTRAÇÃO ===")
```

```python
if __name__ == "__main__":
    demo()
```
